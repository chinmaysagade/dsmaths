### Matplotlib
matplotlib is a Python-based plotting library with full support for 2D and limited support for 3D graphics, extensively used for data analysis and visualization. It was created by John Hunter while he and his colleagues were on epilepsy research as a visualizing tool for electroencephalography and electrocorticography analysis.   
#### Architecture
The current matplotlib architecture revolves around the operations that are necessary for the users to create, render, and update the **Figure** objects.Figure is the top-level matplotlib object that contains and manages all of the elements in a given graphic.  
The architecture to accomplish this is logically separated into three layers, which can be viewed as a stack. Each layer that sits above another layer knows how to talk to the layer below it, but the lower layer is not aware of the layers above it.

<img src="images/matplotlib_layers.jpg",width=600,height=600,align="left" style="float:left;margin-right:2%">

**Backend Layer:**   Provides concrete implementations of the drawing surface , renderer and even handlers (for user inputs and zooming etc) .  
**Artist Layer:**   This is the layer where the plots are generated and consists of classes such as lines, shapes, axes, text, and so on. The objects in artist layer know how to use renderers and draw on the surface.  Everything you see in a matplotlib figure is an artist instance.   
**Scripting Layer (pyplot):**  Scripting layer is the user-facing interface that simplifies the task of working with other layers.  

**rcParams**
rcParams is  global dictionary of default configuration settings. It is initialized by code which may be overridded by a matplotlibrc file.

#### Anatomy of matplot lib figure

<img src="images/anatomy1.png",width=600,height=600,align="left" style="float:left;margin-right:2%">

**Figure:**  Whole plot area containing all the elements.  
**Grid:** To tile multiple subplots in one figure.  
**Axis:** Measurement of value of a point in a given dimension. Normally, you would have a X and a Y axis.  
**Axes:** The plural form of axis represents the data plotting area.  
**Title:** The name of the figure.  
**Spines:** The lines that denote the boundaries of the axes.  
**Axis-Label:** Description of axis.  
**Ticks:** Marks of division on axis.  
**Tick-Labels:** values of major and minor ticks.  
**Legend:** Labels for each data series.  
**Patches:** Shape added by matplotlib. Rectangles, circles, polygons etc.

#### Note on Marker Size
Marker size property is different for different plots.  
For plot function, use ms or markersize property, like ms=10.  
For scatter function, use property 's' like shown below.  S is usually square of the marker size property of the plot function.  
https://stackoverflow.com/questions/14827650/pyplot-scatter-plot-marker-size

```python
import numpy as np
x1= np.random.rand(20)
y1= np.random.rand(20)
x2= np.random.rand(20)
y2= np.random.rand(20)
plt.scatter(x1,y1,label="P1",marker="+",s=100)
plt.scatter(x2,y2,label="P2",marker="^",s=120)
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.show() 
```

#### Grids
Grid lines extend from tick marks. To add a grid, call plt.grid function.

#### Ticks
Ticks are the dividers on the axis. The can also be used to label data series like in barcharts.  
Tick spacing can be set by calling set_major_locator and set_minor_locator api's on the axis objects. Hence, we need to get the axis first to update ticks.  
**Some methods are on plot object while others are on axis. **

```python
x1= [1,2,3]
y1= [10,20,30]

ax=plt.gca() # gca stands for get current axis
tick_labels=['Cat','Dog','Horse']
plt.bar(x1,y1,label="P1",tick_label=tick_labels)
plt.xticks(rotation=25)
ax.xaxis.set_tick_params(labelsize=15)
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.show() 
```

### Adjusting layout
To adjust the size of the figure, directly call plt.figure(figsize=(width-inches,height-inches))

```python
matplotlib.rcParams.update({'font.size':10,'font.family':'serif'})
plt.figure(figsize=(2,2))
squares=[x**2 for x in range(10)]
plt.plot(numbers,squares,label="squares")
plt.legend(loc=3)
plt.title("Squares")
plt.savefig('square1.png',dpi=300)
plt.show() 
```

```python
plt.figure(figsize=(2,2))
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
squares=[x**2 for x in range(10)]
plt.plot(numbers,squares,label="squares")
plt.legend(loc=3)
plt.title("Squares")
plt.savefig('square1.png',dpi=300)
plt.show() 
```

### Subplots
plt.subplots(nrows,ncols,plot_number)  
for rows and cols less than 10, we can use shorthand notation, 111 means 1st row, 1st column and 1st figure.  

```python
squares=[x**2 for x in range(10)]
cubes=[x*x*x for x in range(10)]
figure,axarr=plt.subplots(1,2,sharex=True,sharey=True)
axarr[0].plot(numbers,numbers,label="linear",c="r")
axarr[1].plot(numbers,squares,label="squares",c="b")
plt.show() 
```

#### Inset Plots

```python
squares=[x**2 for x in range(10)]
cubes=[x*x*x for x in range(10)]
figure,ax1=plt.subplots()
left,bottom,width,height = [0.22,0.45,0.3,0.35]
ax2=figure.add_axes([left,bottom,width,height])
ax1.plot(numbers,numbers,label="linear",c="r")
ax2.plot(numbers,squares,label="squares",c="b")
plt.show() 
```
