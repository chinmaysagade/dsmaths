```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Linear Function:

```python
t = np.arange(1, 10, 1)
s = 1+0.5*t
plt.plot(t,s)
plt.title(r'Linear Function: 1+0.5*t', fontsize=20)
plt.show()
```

Power Function:

```python
t = np.arange(1, 10, 1)
s = t**2+t+1
plt.plot(t,s)
plt.title(r'$Power Function: t^2+t+1$', fontsize=20)
plt.show()
```

Exponential Function:

```python
t = np.arange(1, 10, 1)
s = 2**t+t+1
plt.plot(t,s)
plt.title(r'$Exponential \ Function: 2^t+t+1$', fontsize=20)
plt.show()
```

Logarithmic Function:

```python
t = np.arange(1, 10, 1)
s = np.log(t)+1
plt.plot(t,s)
plt.title(r'$Logarithmic \ Function: \log_et+1$', fontsize=20)
plt.show()
```

Sigmoid Function:

```python
t = np.arange(-10, 10, 1)
s = np.exp(t)/(np.exp(t)+1)
plt.plot(t,s)
plt.title(r'$Logarithmic \ Function: \log_et+1$', fontsize=20)
plt.show()
```
